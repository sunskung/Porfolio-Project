## Combine all sheets into 1 excel in one tab

```python
#  Import packages and set the working directory

import os
import glob
import pandas as pd
os.chdir(r"C:\Users\Phurin Chawjarusopas\Desktop\Agoda Case Study\KPMG Virtual Internship\Cleaned data")
```

```python
# Use glob to match the pattern ‘csv’

extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
all_filenames
```

```python
# Combine all files in the list and export as CSV

#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')
```

```python
import os
import pandas as pd
import glob


path = r'C:\Users\Phurin Chawjarusopas\Desktop\Agoda Case Study\KPMG Virtual Internship\Cleaned data'
all_files = glob.glob(os.path.join(path, "*.csv"))

writer = pd.ExcelWriter('out.xlsx')

for f in all_files:
    df = pd.read_csv(f)
    df.to_excel(writer, sheet_name=os.path.splitext(os.path.basename(f))[0], index=False)

writer.save()
```
